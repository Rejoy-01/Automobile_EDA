##Total count of Product Lines 

```python
import seaborn as sns
import matplotlib.pyplot as plt
usr = df1.groupby('Product_line').size().reset_index(name= 'Size')
sns.barplot(x="Product_line", y="Size", data=usr, palette='hls').set_title('Product Line Count')
# plt.xlabel=('Product Line')
plt.ylabel =('Size')
plt.tick_params(axis='x',rotation=45)
plt.show()
```

Classic cars and Vintage cars are sold more while the the Trains are sold the least
